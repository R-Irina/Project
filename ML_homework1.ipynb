```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import tree
```

```python
hous_df = pd.read_csv('housing.csv')
hous_df.head()
```

```python
hous_df.describe()
```

```python
hous_df.info()
```

```python
hous_df
```

```python
x = hous_df.drop('MEDV',axis=1)
y = hous_df['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(hous_df.drop(['MEDV'],axis=1), 
                                                    hous_df['MEDV'], 
                                                    test_size=0.2, 
                                                    random_state=106)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
print('Ошибка MSE=',mean_squared_error(y_test,pred,squared=True))
```

```python
plt.figure(figsize=(20,9))
plt.plot(y_test.values, "blue", linewidth=5, label = 'Реальные значения')
plt.plot(pred, "red", linewidth=5, label = 'Наш первый ML')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

Decision Tree Regressor

```python
X_train, X_test, y_train, y_test = train_test_split(hous_df.drop('MEDV',axis=1), 
                                                    hous_df['MEDV'], 
                                                    test_size=0.2, 
                                                    random_state=286)
```

```python
tree_regressor = tree.DecisionTreeRegressor(max_depth=10)
tree_regressor.fit(X_train,y_train)
pred_tree = tree_regressor.predict(X_test)
```

```python
print('Ошибка MSE Decision TreeRegressor=',mean_squared_error(y_test,pred_tree))
```

```python
plt.figure(figsize=(20,9))
plt.plot(y_test.values, "blue", linewidth=5, label = 'Реальные значения')
plt.plot(pred_tree, "red", linewidth=5, label = 'ML TreeRegressor')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
frame = {'Важность параметров':tree_regressor.feature_importances_,'Параметры':X_train.columns}
result = pd.DataFrame(frame)
result.sort_values(by='Важность параметров', ascending=False)
```
